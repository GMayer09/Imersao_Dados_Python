```python
df.columns
```

```python
column_mapping = {
    'work_year': 'ano_trabalho',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_funcionario',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=column_mapping)
display(df.head())
```

```python
df['nivel_experiencia'].value_counts()
```

```python
df["tipo_emprego"].value_counts()
```

```python
df["taxa_remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
df['nivel_experiencia'] = df['nivel_experiencia'].replace({
    'SE': 'Snior',
    'MI': 'Pleno',
    'EN': 'Jnior',
    'EX': 'Executivo'
})

df['nivel_experiencia'].value_counts()
```

```python
df['tipo_emprego'] = df['tipo_emprego'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Perodo',
    'FL': 'Freelancer'
})

df['tipo_emprego'].value_counts()
```

```python
df['taxa_remoto'] = df['taxa_remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hbrido'
})

df['taxa_remoto'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
})

df['tamanho_empresa'].value_counts()
```
